```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)
x = np.linspace(0, 100, 100)
noise = np.random.normal(0, 5, size=100)
y = 0.5 * x + 10 + noise

plt.figure(figsize=(7,4))
plt.scatter(x, y, label='Datos', s=20)      # puntos
plt.plot(x, 0.5*x + 10, color='orange', label='Tendencia (sin ruido)')  # l√≠nea ideal
plt.title('Lineal con ruido')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
