```python
# Celda 1
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)           # para que el ejemplo sea reproducible
x = np.arange(100)          # 0,1,2,...,99
y = 200 + np.random.normal(0, 1, size=100)  # ruido alrededor de 200

plt.figure(figsize=(6,4))   # tamao de la imagen en pulgadas
plt.fill_between(x, y, 195, alpha=0.6)  # rea bajo la curva (hasta y=195)
plt.plot(x, y)              # lnea encima del rea
plt.title('Sample Visualization')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```
